```python
import os
import pandas as pd
import toothlib

cfg = toothlib.config

# load data
STATIC = 'static'
edge_df = pd.read_csv(os.path.join(STATIC, cfg['dataset']['processed']['edge_csv']))
p_edge_df = edge_df.iloc[101:200]
```

```python
# plot
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1.5)

# 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 12))
sns.lineplot(x='v', y='gt', data=p_edge_df, ax=axes[0])
sns.lineplot(x='metric_u', y='gt', data=p_edge_df, ax=axes[1])
sns.lineplot(x='curv', y='gt', data=p_edge_df, ax=axes[2])
plt.show()
```

```python
df = pd.read_csv(os.path.join(STATIC, cfg['dataset']['processed']['depth_csv']))

p1_df = df['curv'].iloc[0:10000]
p2_df = df['curv'].iloc[20000:30000]
vmin = min(p1_df.min(), p2_df.min())
vmax = max(p1_df.max(), p2_df.max())

# not axis
sns.heatmap(p1_df.values.reshape(100, 100), vmin=vmin, vmax=vmax, cmap='YlGnBu', cbar=False, square=True, xticklabels=False, yticklabels=False)
plt.show()
sns.heatmap(p2_df.values.reshape(100, 100), vmin=vmin, vmax=vmax, cmap='YlGnBu', cbar=False, square=True, xticklabels=False, yticklabels=False)
plt.show()
```
